```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[11,23,43,52,57,67,76]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color='red',linewidth=5,linestyle='dashdot')
```

```python
plt.plot(x,y,color='red',linewidth=5,linestyle='dotted')
```

```python
plt.xlabel('cost')
plt.ylabel('profit')
plt.title('small gain')
plt.plot(x,y,color='yellow',linewidth=5,linestyle='dotted')
```
