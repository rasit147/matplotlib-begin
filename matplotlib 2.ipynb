# Axes labels,legend, grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')
plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")
plt.legend(loc="best")  
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')
plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")
plt.legend(loc="best",shadow=True, fontsize="small")  
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather')
plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")
plt.legend(loc="best",shadow=True, fontsize="small")  
plt.grid()
```

# Histogram

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar,bins=3,rwidth=0.95)
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel('sugar range')
plt.ylabel("Total no.of patient")
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','red'],label=['man','women'])
plt.legend()
plt.grid()
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','red'],label=['man','women'],orientation='horizontal')
```

# pie chart 

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%',shadow=True)
plt.show()
```

```python
# explode function 
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%',shadow=True,explode=[0,0.1,0.1,0,0])
plt.show()
```

```python
#startangle 
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%',shadow=True,explode=[0,0.1,0.1,0,0],startangle=270)
plt.show()
plt.savefig("piechart.png")
```
